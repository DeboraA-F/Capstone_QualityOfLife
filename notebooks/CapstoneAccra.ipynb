```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
```

## Cost of Living

```python
URL = 'https://www.numbeo.com/cost-of-living/in/Accra'
response = requests.get(URL)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
soup.find('table', attrs={'class':'data_wide_table new_bar_table'})
```

```python
cost_of_living_accra = pd.read_html(str(soup.find('table', attrs={'class':'data_wide_table new_bar_table'})))[0]
```

```python
cost_of_living_accra
```

```python
cost_of_living_accra.to_csv('Accra_Cost_of_Living.csv', index=False)
```

## Crime

```python
URL = 'https://www.numbeo.com/crime/in/Accra'
response = requests.get(URL)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
soup.find('table', attrs={'class':'table_builder_with_value_explanations data_wide_table'})
```

```python
crime = pd.read_html(str(soup.find('table', attrs={'class':'table_builder_with_value_explanation data_wide_table'})))[0]
```

```python
crime
```

```python
crime = crime.drop(columns = [1])
```

```python
crime = crime.rename(columns = {0 : 'Types of Crime', 2: 'Crime%', 3: 'Value'})
crime
```

```python
crime.to_csv('Accra_crime.csv', index=False)
```

SAFETY IN ACCRA

```python
safety = pd.read_html(str(soup.findAll('table', attrs={'class':'table_builder_with_value_explanation data_wide_table'})))[1]
```

```python
safety = safety.drop(columns = [1])
```

```python
safety = safety.rename(columns = {0 : 'Components of Safety', 2: 'Safety%', 3: 'Value'})
safety
```

```python
safety.to_csv('Accra_safety.csv', index=False)
```

## Health Care

```python
URL = 'https://www.numbeo.com/health-care/in/Accra'
response = requests.get(URL)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
soup.find('table', attrs={'class':'table_builder_with_value_explanation data_wide_table'})
```

```python
health_care = pd.read_html(str(soup.find('table', attrs={'class':'table_builder_with_value_explanation data_wide_table'})))[0]
```

```python
health_care = health_care.drop(columns = ['Satisfaction %'])
health_care
```

```python
health_care = health_care.rename(columns = {'Satisfaction %.1' : 'Satisfaction %', 'Unnamed: 3': 'Value'})
```

```python
health_care.to_csv('Accra_health_care.csv', index=False)
```
